```python
url = 'https://www.federalreserve.gov/mediacenter/files/FOMCpresconf20200129.pdf'

r = requests.get(url)
f = io.BytesIO(r.content)
returnstring = []
reader = PdfFileReader(f)
for i in range(reader.getNumPages()):
    contents = reader.getPage(i).extractText().replace('\n', " ")
    returnstring.append(contents)
returnstring
```
